# Visualizing Washington DC Bike Share

Looking at the relationship between humidity levels and total rentals to see if there is an interesting relationship we might want to explore later.

```
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('ggplot')
```

```
df = pd.read_csv('https://assets.datacamp.com/production/repositories/2210/datasets/fb4f2c1039e3df2c2e2624a8c95de5a1980861c6/bike_share.csv')
df.head()
```

```
sns.set_style('whitegrid')
g = sns.JointGrid(data=df, x='hum', y='total_rentals', xlim=(0.1,1.0), height=13)
g.plot(sns.regplot, sns.distplot)
plt.show()
```

```
sns.jointplot(data=df, x='hum', y='total_rentals', kind='reg', height=13)
plt.show()
```

 These plots show that there is limited relationship between rental amounts and humidity levels.

 # temp vs. total_rentals

```
sns.jointplot(data=df, kind='reg', x='temp', y='total_rentals', xlim=(0,1),order=2, height=13)
plt.show()
```

```
sns.jointplot(data=df, kind='resid', x='temp', y='total_rentals', xlim=(0,1),order=2, height=13)
plt.show()
```

```
g = (sns.jointplot(x="temp",
             y="casual",
             kind='scatter',
             data=df,height=13,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
plt.show()
```

```
g = (sns.jointplot(x="temp",
             y="registered",
             kind='scatter',
             data=df,height=13,
             marginal_kws=dict(bins=10, rug=True))
    .plot_joint(sns.kdeplot))
plt.show()
```
